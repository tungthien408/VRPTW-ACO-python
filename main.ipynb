```python
import pandas as pd
import matplotlib.pyplot as plt

from vrptw_base import VrptwGraph
from basic_aco import BasicACO
```

```python

file_path = './dataset/homberger_400_customer_instances/C1_4_1.txt'
result_path = './results/param4/'
filename = 'aco_results_400.csv'

ants_num = 20
max_iter = 200
beta = 2
q0 = 0.1
show_figure = False 

graph = VrptwGraph(file_path)
basic_aco = BasicACO(graph, ants_num=ants_num, max_iter=max_iter, beta=beta, q0=q0, result_path=result_path, filename=filename,
                     whether_or_not_to_show_figure=show_figure)

basic_aco.run_basic_aco()
```

```python
result_file = './results/param4/aco_results_400.csv'
df = pd.read_csv(result_file)

plt.figure(figsize=(12, 6))

plt.plot(df['iteration'], df['best_cost'], 
         linewidth=2, color='blue', label='Best Cost')

plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Cost (Distance)', fontsize=12)
plt.title('ACO Convergence - Best Cost over Iterations', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
